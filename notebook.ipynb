## Menentukan Pertanyaan Bisnis

- Stasiun pengamatan manakah yang memiliki rata-rata konsentrasi suatu polutan udara yang paling tinggi?

Pertanyaan ini bertujuan untuk mengidentifikasi stasiun pengamatan yang memiliki rata-rata konsentrasi tertinggi untuk polutan tertentu (misalnya PM2.5, PM10, SO2, NO2, CO, atau O3). Informasi ini sangat berguna untuk beberapa tujuan:

Mengidentifikasi Hotspot Polusi: Dengan mengetahui stasiun mana yang memiliki konsentrasi polutan tertinggi, pihak berwenang dapat mengidentifikasi daerah yang paling terpengaruh oleh polusi udara. Hal ini penting untuk perencanaan kebijakan dan tindakan pengendalian polusi.

Perencanaan Intervensi: Data ini bisa digunakan untuk merencanakan intervensi yang lebih fokus di area-area yang paling membutuhkan. Misalnya, peningkatan area hijau atau pengaturan lalu lintas di area dengan konsentrasi polutan tinggi.

Pemahaman Polusi Lokal: Analisis ini membantu memahami variasi geografis dalam polusi udara dan bagaimana faktor-faktor lokal seperti aktivitas industri, lalu lintas, atau kondisi cuaca mempengaruhi kualitas udara.


- Apakah ada korelasi antara polutan PM2.5 dengan rata-rata polutan lain?

Pertanyaan ini bertujuan untuk mengeksplorasi hubungan antara konsentrasi PM2.5 dengan polutan lainnya (seperti PM10, SO2, NO2, CO, atau O3). Korelasi ini dapat memberikan wawasan yang berguna dalam beberapa cara:

Identifikasi Polutan Bersama: Jika terdapat korelasi yang kuat antara PM2.5 dan polutan lainnya, ini mungkin menunjukkan bahwa polutan tersebut memiliki sumber yang sama atau dipengaruhi oleh kondisi lingkungan yang serupa. Misalnya, aktivitas industri atau lalu lintas bisa menghasilkan berbagai jenis polutan secara bersamaan.

Pemahaman Mekanisme Polusi: Memahami korelasi antar polutan membantu dalam memahami mekanisme pencemaran udara. Misalnya, NO2 dan O3 sering menunjukkan hubungan yang kompleks di atmosfer, di mana reaksi kimia tertentu dapat meningkatkan atau menurunkan konsentrasi masing-masing.

Perencanaan Kesehatan dan Kebijakan: Korelasi antara polutan bisa digunakan untuk perencanaan kesehatan dan kebijakan. Jika beberapa polutan menunjukkan hubungan yang erat, kebijakan yang mengurangi salah satu polutan mungkin juga memiliki efek mengurangi polutan lainnya, memberikan manfaat kesehatan ganda.

Pertanyaan bisnis yang ingin dijawab dengan menggunakan analisis data:

```python
all_df = pd.concat([aotizhongxin_df, changping_df, dingling_df, dongsi_df, guanyuan_df, gucheng_df, huairou_df, nongzhanguan_df, shunyi_df, tiantan_df, wanliu_df, wanshouxigong_df], ignore_index=True)
```

```python
all_df.describe()
```

Gunakan fungsi groupby() dan agg() untuk mengelompokkan seluruh data berdasarkan stasiun dan menghitung nilai max dan rata-rata dari setiap polutan yang dicari seperti PM2.5, PM10, NO2, SO2, CO, dan O3



```python
all_df.groupby(by='station').agg({
    "PM2.5":['max', 'mean'],
    "PM10":['max', 'mean'],
    "NO2":['max', 'mean'],
    "SO2":['max', 'mean'],
    "CO":['max', 'mean'],
    "O3":['max', 'mean']
})
```

Membuat DataFrame baru dengan label pollutant_all_df untuk melihat korelasi antara masing-masing polutan

```python
pollutant_all_df = all_df[['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']]
```

```python
pollutant_all_df.corr()
```

- Terdapat korelasi yang kuat antara PM2.5 dengan PM10 dan CO, menunjukkan bahwa polutan ini cenderung meningkat bersama-sama.
- Terdapat korelasi yang cukup kuat antara PM2.5 dengan NO2, menunjukkan adanya hubungan signifikan antara kedua polutan ini.
- Korelasi antara PM2.5 dengan SO2 bersifat moderat, menunjukkan hubungan yang ada namun tidak terlalu kuat.
- Korelasi antara PM2.5 dengan O3 bersifat negatif lemah, menunjukkan adanya hubungan terbalik yang lemah antara kedua polutan ini.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun pengamatan manakah yang memiliki rata-rata konsentrasi suatu polutan udara yang paling tinggi?

Untuk keperluan visualisasi, saya membuat DataFrame terpisah untuk setiap polutan. Setiap DataFrame dibuat dengan mengelompokkan data berdasarkan nama stasiun menggunakan fungsi `groupby()` dan `agg()`, kemudian diurutkan berdasarkan nilai polutan tertinggi dengan menambahkan parameter `ascending=False` pada fungsi `sort_values()`. Fungsi `reset_index()` digunakan untuk mengatur ulang DataFrame sehingga tidak memiliki indeks.

```python
pollutant_pm25 = all_df.groupby(by='station').agg({"PM2.5":'mean'}).sort_values(by='PM2.5', ascending=False).reset_index()
pollutant_pm25.rename(columns={
    'station':'Distrik',
    'PM2.5':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_pm10 = all_df.groupby(by='station').agg({"PM10":'mean'}).sort_values(by='PM10', ascending=False).reset_index()
pollutant_pm10.rename(columns={
    'station':'Distrik',
    'PM10':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_so2 = all_df.groupby(by='station').agg({"SO2":'mean'}).sort_values(by='SO2', ascending=False).reset_index()
pollutant_so2.rename(columns={
    'station':'Distrik',
    'SO2':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_no2 = all_df.groupby(by='station').agg({"NO2":'mean'}).sort_values(by='NO2', ascending=False).reset_index()
pollutant_no2.rename(columns={
    'station':'Distrik',
    'NO2':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_co = all_df.groupby(by='station').agg({"CO":'mean'}).sort_values(by='CO', ascending=False).reset_index()
pollutant_co.rename(columns={
    'station':'Distrik',
    'CO':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_o3 = all_df.groupby(by='station').agg({"O3":'mean'}).sort_values(by='O3', ascending=False).reset_index()
pollutant_o3.rename(columns={
    'station':'Distrik',
    'O3':'Rata-Rata Konsentrasi'
}, inplace=True)
```

Memvisualisasikan data dengan bar plot untuk melihat stasiun pengamatan mana saja yang memiliki konsentrasi polutan tertinggi untuk masing-masing polutan dengan memvisualisasikan 5 stasiun pengamatan teratas. Dengan menggunakan pallet warna yang berbeda, stasiun dengan konsentrasi polutan paling tinggi dapat disorot dengan warna yang berbeda dan dapat dicermati dengan mudah.

```python
import matplotlib.pyplot as plt
# Visualisasi PM2.5
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_pm25[:5], palette='Blues_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi PM2.5 Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi PM2.5', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi PM10
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_pm10[:5], palette='Greens_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi PM10 Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi PM10', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi SO2
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_so2[:5], palette='Oranges_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi SO2 Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi SO2', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi NO2
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_no2[:5], palette='Reds_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi NO2 Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi NO2', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi CO
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_co[:5], palette='Purples_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi CO Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi CO', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi O3
plt.figure(figsize=(10, 6))
sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_o3[:5], palette='YlOrBr_r')
plt.title('5 Stasiun dengan Rata-Rata Konsentrasi O3 Tertinggi', fontsize=15, pad=20)
plt.xlabel('Rata-Rata Konsentrasi O3', fontsize=12)
plt.ylabel('Distrik', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
